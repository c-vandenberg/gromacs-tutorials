```python
import numpy
import matplotlib.pyplot as pyplot
```

# Extract time and potential energy from `potential.xvg` and convert to picoseconds (based on emstep in `emin-charmm.mdp` force-field used in energy minimization)

```python
time, potential_energy = numpy.loadtxt(
        '../potential.xvg'
    ).T

# Multiply time by emstep to get time in picoseconds
time *= 0.01
```

# Create Matplotlib graph to plot potential energy vs time

```python
# Set up the line graph
potential_energy_vs_time_line_graph, potential_energy_vs_time_line_graph_axes = pyplot.subplots(figsize=(18, 6))
potential_energy_vs_time_line_graph.patch.set_facecolor('black')
potential_energy_vs_time_line_graph_axes.set_facecolor('black')

# Stylise ticks and spines
potential_energy_vs_time_line_graph_axes.tick_params(colors='white', which='both', labelsize=20)
for spine in potential_energy_vs_time_line_graph_axes.spines.values():
    spine.set_edgecolor('white')

# Plot the data
potential_energy_vs_time_line_graph_axes.plot(time, potential_energy, color='cyan', linewidth=2)

# Set axes labels, limits and graph title
potential_energy_vs_time_line_graph_axes.set_xlabel('$t$ (ps)', fontsize=20, color='white')
potential_energy_vs_time_line_graph_axes.set_ylabel(r'$Potential\ Energy$ (KJ/mol)', fontsize=20, color='white')
potential_energy_vs_time_line_graph_axes.set_xlim(0, 10)
potential_energy_vs_time_line_graph_axes.set_ylim(-630000, -155000)
potential_energy_vs_time_line_graph_axes.set_title(r'$\bf{Protein\ 1FJS\ Potential\ Energy\ vs\ Time}$', 
                                             fontsize=20, color='white')

# Add figure title
potential_energy_vs_time_line_graph.text(0.5, -0.075,
                r'$\bf{Fig\ 1}$ Evolution of protein 1FJS average potential energy as a function of time '
                r'during steepest descent minimization',
                ha='center', va='center', color='white', fontsize=20)

# Save & display the plot
pyplot.savefig('../../reports/figures/graphs/potential-energy-vs-time.png', bbox_inches='tight')
pyplot.show()
```
