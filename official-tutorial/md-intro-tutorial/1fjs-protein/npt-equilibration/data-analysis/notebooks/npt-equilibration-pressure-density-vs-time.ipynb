```python
import numpy
import matplotlib.pyplot as pyplot
```

# Extract time, pressure and density from `pressure.xvg` and `density.xvg`

```python
time, pressure = numpy.loadtxt(
        '../pressure.xvg'
    ).T

_, density = numpy.loadtxt(
        '../density.xvg'
    ).T
```

# Create Matplotlib graph to plot pressure vs time

```python
# Set up the line graph
pressure_vs_time_line_graph, pressure_vs_time_line_graph_axes = pyplot.subplots(figsize=(18, 6))
pressure_vs_time_line_graph.patch.set_facecolor('black')
pressure_vs_time_line_graph_axes.set_facecolor('black')

# Stylise ticks and spines
pressure_vs_time_line_graph_axes.tick_params(colors='white', which='both', labelsize=20)
for spine in pressure_vs_time_line_graph_axes.spines.values():
    spine.set_edgecolor('white')

# Plot the data
pressure_vs_time_line_graph_axes.plot(time, pressure, color='cyan', linewidth=3.5)

# Set axes labels, limits and graph title
pressure_vs_time_line_graph_axes.set_xlabel('$t$ (ps)', fontsize=20, color='white')
pressure_vs_time_line_graph_axes.set_ylabel(r'$Pressure$ (Bar)', fontsize=20, color='white')
pressure_vs_time_line_graph_axes.set_xlim(0, 100)
pressure_vs_time_line_graph_axes.set_ylim(-5250, 500)
pressure_vs_time_line_graph_axes.set_title(r'$\bf{Protein\ 1FJS\ Pressure\ vs\ Time}$', fontsize=20, color='white')

# Add figure title
pressure_vs_time_line_graph.text(0.5, -0.075,
                r'$\bf{Fig\ 1}$ Evolution of protein 1FJS pressure as a function of time during NVT ensemble '
                r'equilibration',
                ha='center', va='center', color='white', fontsize=20)

# Save & display the plot
pyplot.savefig('../../reports/figures/graphs/pressure-vs-time.png', bbox_inches='tight')
pyplot.show()
```

# Create Matplotlib graph to plot density vs time

```python
density_vs_time_line_graph, density_vs_time_line_graph_axes = pyplot.subplots(figsize=(18, 6))
density_vs_time_line_graph.patch.set_facecolor('black')
density_vs_time_line_graph_axes.set_facecolor('black')

# Stylise ticks and spines
density_vs_time_line_graph_axes.tick_params(colors='white', which='both', labelsize=20)
for spine in density_vs_time_line_graph_axes.spines.values():
    spine.set_edgecolor('white')

# Plot the data
density_vs_time_line_graph_axes.plot(time, density, color='cyan', linewidth=3.5)

# Set axes labels, limits and graph title
density_vs_time_line_graph_axes.set_xlabel('$t$ (ps)', fontsize=20, color='white')
density_vs_time_line_graph_axes.set_ylabel(r'$Density$ (kg/m$^{3}$)', fontsize=20, color='white')
density_vs_time_line_graph_axes.set_xlim(0, 100)
density_vs_time_line_graph_axes.set_ylim(1000, 1050)
density_vs_time_line_graph_axes.set_title(r'$\bf{Protein\ 1FJS\ Density\ vs\ Time}$',
                                             fontsize=20, color='white')

# Add figure title
density_vs_time_line_graph.text(0.5, -0.075,
                r'$\bf{Fig\ 2}$ Evolution of protein 1FJS density as a function of time during NVT ensemble '
                r'equilibration',
                ha='center', va='center', color='white', fontsize=20)

# Save & display the plot
pyplot.savefig('../../reports/figures/graphs/density-vs-time.png', bbox_inches='tight')
pyplot.show()
```
