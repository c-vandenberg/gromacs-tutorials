```python
import numpy
import matplotlib.pyplot as pyplot
```

# Extract time and temperature from temperature.xvg and zero-index time

```python
time, temperature = numpy.loadtxt(
        '../temperature.xvg'
    ).T

# Subtracts the first element from every element in the time array, ensuring the first element is 0/the time starts
# at zero
time -= time[0]
```

# Create Matplotlib graph to plot temperature vs time

```python
# Set up the line graph
temperature_vs_time_line_graph, temperature_vs_time_line_graph_axes = pyplot.subplots(figsize=(18, 6))
temperature_vs_time_line_graph.patch.set_facecolor('black')
temperature_vs_time_line_graph_axes.set_facecolor('black')

# Stylise ticks and spines
temperature_vs_time_line_graph_axes.tick_params(colors='white', which='both', labelsize=20)
for spine in temperature_vs_time_line_graph_axes.spines.values():
    spine.set_edgecolor('white')

# Plot the data
temperature_vs_time_line_graph_axes.plot(time, temperature, color='cyan', linewidth=3.5)

# Set axes labels, limits and graph title
temperature_vs_time_line_graph_axes.set_xlabel('$t$ (ps)', fontsize=20, color='white')
temperature_vs_time_line_graph_axes.set_ylabel(r'$Temperature$ (K)', fontsize=20, color='white')
temperature_vs_time_line_graph_axes.set_xlim(0, 80)
temperature_vs_time_line_graph_axes.set_ylim(295, 305)
temperature_vs_time_line_graph_axes.set_title(r'$\bf{Protein\ 1FJS\ Temperature\ vs\ Time}$', fontsize=20, color='white')

# Add figure title
temperature_vs_time_line_graph.text(0.5, -0.075,
                r'$\bf{Fig\ 1}$ Evolution of protein 1FJS temperature as a function of time during NVT ensemble '
                r'equilibration',
                ha='center', va='center', color='white', fontsize=20)

# Save & display the plot
pyplot.savefig('../../reports/figures/graphs/temp-vs-time.png', bbox_inches='tight')
pyplot.show()
```
